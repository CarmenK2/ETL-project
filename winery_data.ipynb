### Transformation Wine Volume by Region

```python
# Display dataframe
GIVolumeProduction_df.head()
```

```python
# Drop 'Volume % change on previous periode' and 'Average value % change on previous period'
GIVolumeProductionClean_df = GIVolumeProduction_df[["Geographical Indication", "Year", "Volume (L)", "Average value"]]

# Display dataframe
GIVolumeProductionClean_df.head()
```

### Transformation Wine Sales by Grape Variety
* Rename column
* Drop all rows with NaN value

```python
# Display dataframe
SalesByVariety_df.head()
```

```python
# Rename 'Variety' to 'Grape Variety', add 'Export Sales' to each year
SalesByVariety_df = SalesByVariety_df.rename(columns={"Variety":"Grape Variety", "2021":"2021 Export Sales", "2020":"2020 Export Sales", "2019":"2019 Export Sales", "2018":"2018 Export Sales", "2017":"2017 Export Sales", "2016":"2016 Export Sales", "2015":"2015 Export Sales", "2014":"2014 Export Sales", "2013":"2013 Export Sales", "2012":"2012 Export Sales"})

# Display dataframe
SalesByVariety_df.head()
```

### Transformation Grape Variety Production
* Replace NaN with 0
* Drop '% change in average value' column
* Rename 'Variety' to 'Grape Variety', add 'Export Sales' to each year

```python
# Rename 'Variety' to 'Grape Variety', add 'Export Sales' to each year
WineProductionSummary_df = WineProductionSummary_df.rename(columns={"Colour/Variety":"Wine Colour", "Category":"Grape Variety", "Winery grown fruit":"Winery grown fruit per tonne", "Total crushed":"Total crushed (T)"})

# Display dataframe
WineProductionSummary_df.head()
```
