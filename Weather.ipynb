```python
#importing the weather station 9025 rainfall
csv_file_9025_rainfall = "Resources_weather/9025_rainfall.csv"
rainfall_9025_data_df = pd.read_csv(csv_file_9025_rainfall)
rainfall_9025_data_df
```

```python
#Fill all the NA value with Zero
rainfall_9025_fillna_df = rainfall_9025_data_df.fillna(00)
rainfall_9025_fillna_df
```

```python
#Correct the sumation for the Annual column after filling all NA value with Zero
rainfall_9025_fillna_df['Annual']=rainfall_9025_fillna_df[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']].sum(axis=1)
```

```python
rainfall_9025_fillna_df
```

```python
#Groupby the Station Number for the average rainfall since year 2000
rainfall_9025_agg=rainfall_9025_fillna_df.loc[rainfall_9025_fillna_df["Year"]>=2000,["Station Number","Annual"]]
rainfall_9025_agg.groupby("Station Number").mean()
```

Creating a consolidated master file to group all the rainfall csv file into one and then apply the data cleaning and transformation

```python
#Import os to interact with the underlying operating system/ returns the absolute path of the working directory
import os
os.getcwd()
```

```python
#Looping the Resources_weather folder for each file name under rainfall.csv and join them together into a master file

dfs=[]

for file in os.listdir("Resources_weather"):
   

    if file.endswith("rainfall.csv"):
        dfs.append(pd.read_csv(f"Resources_weather/{file}"))
        print(file)



```

```python
rainfall_df=pd.concat(dfs)
rainfall_df.head()
```

```python
#Fill up all the NA value with Zero value and correct the summation of "Annual" column 

rainfall_clean_df = rainfall_df.fillna(00)

rainfall_clean_df['Annual']=rainfall_clean_df[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']].sum(axis=1)
rainfall_clean_df
```

```python
#Selecting only year older than 2000, and column of station number and annual
rainfall_data_df=rainfall_clean_df.loc[rainfall_clean_df["Year"]>=2000,["Station Number","Annual"]]
rainfall_data_df
```

The weather station 12280 has faulty data, so will use the substitute weather station 12304 which we downloaded as a backup previously. 
Weather station 12304 is 2nd closest station for the Lat / Lng location

```python
#Drop off weather station 12280 as data are faulty
rainfall_data_excl_df = rainfall_data_df[rainfall_data_df["Station Number"] != 12280]
rainfall_data_excl_df
```

```python
#check how many station left
No_of_station = len(rainfall_data_excl_df["Station Number"].unique())
No_of_station
```

```python
#Name change for "Annual Column" to "Annual Average Rainfall (mm)"
rainfall_data_namechange_df=rainfall_data_excl_df.rename(columns={"Annual":"Annual Average Rainfall (mm)"})
rainfall_data_namechange_df
```

```python
#Groupby the station number and calculate the mean for year older than 2000
rainfall_data_average_df=rainfall_data_namechange_df.groupby("Station Number").mean()
rainfall_data_average_df
```

```python
#Round to 2 decimal place for better presentation
rainfall_data_average_FINAL_df=rainfall_data_average_df.round({'Annual Average Rainfall (mm)':2})
rainfall_data_average_FINAL_df
```

```python
#Export to csv file
rainfall_data_average_FINAL_df.to_csv("Resources_weather/rainfall.csv", index=True)
```
