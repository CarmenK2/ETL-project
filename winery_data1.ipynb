### Transformation Wine Sales by Geographical Location

```python
# Display Dataframe
GeoWineSales_df.head()
```

```python
# Rename 'Label geographical location' to 'Region', add 'Export Sales' to each year
GeoWineSales_df = GeoWineSales_df.rename(columns={"Label geographical indicator":"Region", "2021":"2021_Export_Sales", "2020":"2020_Export_Sales", "2019":"2019_Export_Sales", "2018":"2018_Export_Sales", "2017":"2017_Export_Sales", "2016":"2016_Export_Sales", "2015":"2015_Export_Sales", "2014":"2014_Export_Sales", "2013":"2013_Export_Sales"})

# Display dataframe
GeoWineSales_df.head()
```

```python
# Replace NaN with 0
GeoWineSales_df_df = GeoWineSales_df.fillna(0)

# Display dataframe
GeoWineSales_df_df.head()
```

```python
# Assess dataframe for datatypes
GeoWineSales_df.info()
```

### Transformation Wine Volume by Region
* Drop 'Volume % change on previous period' and 'Average value % change on previous period' columns
* Rename columns

```python
# Display dataframe
GIVolumeProduction_df.head()
```

```python
# Drop 'Volume % change on previous periode' and 'Average value % change on previous period'
GIVolumeProductionClean_df = GIVolumeProduction_df[["Geographical Indication", "Year", "Volume (L)", "Average value"]]

# Display dataframe
GIVolumeProductionClean_df.head()
```

```python
# Rename 'Label geographical location' to 'Region', add 'Export Sales' to each year
GIVolumeProductionClean_df = GIVolumeProductionClean_df.rename(columns={"Geographical Indication":"Region", "Volume (L)":"Volume_per_Liter", "Average value":"Average_Sales_Value_per_Liter"})

# Display dataframe
GIVolumeProductionClean_df.head()
```

```python
# Assess dataframe for datatypes
GIVolumeProductionClean_df.info()
```

### Transformation Wine Sales by Grape Variety
* Rename column
* Drop all rows with NaN value

```python
# Display dataframe
SalesByVariety_df.head()
```

```python
# Rename 'Variety' to 'Grape Variety', add 'Export Sales' to each year
SalesByVariety_df = SalesByVariety_df.rename(columns={"Variety":"Grape_Variety", "2021":"2021_Export_Sales", "2020":"2020_Export_Sales", "2019":"2019_Export_Sales", "2018":"2018_Export_Sales", "2017":"2017_Export_Sales", "2016":"2016_Export_Sales", "2015":"2015_Export_Sales", "2014":"2014_Export_Sales", "2013":"2013_Export_Sales", "2012":"2012_Export_Sales"})

# Display dataframe
SalesByVariety_df.head()
```

```python
# Drop all rows with NaN value
SalesByVariety_df = SalesByVariety_df.dropna()

# Display dataframe
SalesByVariety_df.head()
```

```python
# Assess dataframe for datatypes
SalesByVariety_df.info()
```

### Transformation Grape Variety Production
* Replace NaN with 0
* Drop '% change in average value' column
* Rename 'Variety' to 'Grape Variety', add 'Export Sales' to each year

```python
# Display dataframe
WineProductionSummary_df.head()
```

```python
# Replace NaN with 0
WineProductionSummary_df = WineProductionSummary_df.fillna(0)

# Display dataframe
WineProductionSummary_df.head()
```

```python
# Drop '% change in average value'
WineProductionSummary_df = WineProductionSummary_df[["Region", "Year", "Colour/Variety", "Category", "Tonnes purchased", "Total value purchased", "Average purchase value per tonne", "Winery grown fruit", "Winery grown share of crush", "Total crushed", "Total value all grapes"]]

# Display dataframe
WineProductionSummary_df.head()
```

```python
# Rename 'Variety' to 'Grape Variety', add 'Export Sales' to each year
WineProductionSummary_df = WineProductionSummary_df.rename(columns={"Colour/Variety":"Wine_Colour", "Category":"Grape_Variety", "Tonnes purchased":"Tonnes_purchased", "Total value purchased":"Total_value_purchased", "Average purchase value per tonne":"Average_purchase_value_per_tonne", "Winery grown fruit":"Winery_grown_fruit_per_tonne", "Winery grown share of crush":"Winery_grown_share_of_crush", "Total crushed":"Total_crushed", "Total value all grapes":"Total_value_all_grapes"})

# Display dataframe
WineProductionSummary_df.head()
```

```python
# Assess dataframe for datatypes
WineProductionSummary_df.info()
```

### Load to PostgresSQL
* Connect to local database
* Check for tables
* Use pandas to load csv converted dataframe into database
* Comfirm data has been added by querying the dataframe

#### Load Wine Sales by Geographical Location

```python
# Connect to local database
user_name = "postgres"
password = "postgres"
rds_connection_string = f"{user_name}:{password}@localhost:5432/wa_wine_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check for tables
engine.table_names()
```

```python
# Use pandas to load csv converted dataframe
GeoWineSales_df.to_sql(name='Winery_Export_Sales_by_Geolocation', con=engine, if_exists='append', index=False)
```

```python
# Confirm data has been added by querying the table
pd.read_sql_query('select * from Winery_Export_Sales_by_Geolocation', con=engine).head()
```

```python
# Use pandas to load csv converted dataframe
GIVolumeProductionClean_df.to_sql(name='Wine_Volume_by_region', con=engine, if_exists='append', index=False)
```
